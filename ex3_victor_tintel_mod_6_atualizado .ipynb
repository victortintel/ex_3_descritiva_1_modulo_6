```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

sinasc = pd.read_csv("sinasc_RO_2019.csv")
sinasc
```

```python
sinasc["APGAR5"]
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
sinasc.loc[(sinasc["APGAR5"] >= 0) & (sinasc["APGAR5"] <= 3), "CAT_APGAR5"] = "Asfixia grave"
sinasc.loc[(sinasc["APGAR5"] >= 4) & (sinasc["APGAR5"] <= 5), "CAT_APGAR5"] = "Asfixia moderada"
sinasc.loc[(sinasc["APGAR5"] >= 6) & (sinasc["APGAR5"] <= 7), "CAT_APGAR5"] = "Asfixia leve"
sinasc.loc[(sinasc["APGAR5"] >= 8) & (sinasc["APGAR5"] <= 10), "CAT_APGAR5"] = "Normal"
sinasc
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sns.catplot(x="APGAR5", y="GESTACAO", hue="SEXO", kind="bar", data=sinasc)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tab_cruzada = pd.crosstab(sinasc["GESTACAO"], sinasc["APGAR5"])
tab_cruzada
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
tab_cruzada_perc = tab_cruzada.div(tab_cruzada.sum(axis = 1), axis = 0)
tab_cruzada_perc
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
tab_cruzada_perc.plot.bar()
print("Este conjunto de dados fornecidos pela tabela cruzada acima não corrobora com a afirmação dos médicos porque esses dados eles estão relacionados com o nível de asfixia do bebê relacionado com o tempo de gestação. Provavelmente para saber mais precisamente sobre a afirmação dos médicos devemos ter algum dado relacionado com o dia do parto relacionado como tempo de gestação e não temos isso na tabela cruzada criada.")
```
